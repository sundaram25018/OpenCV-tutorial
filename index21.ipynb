```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
```

```python
blank_w = np.ones((500,500,3),np.uint8)*255
blank_b = np.zeros((500,500,3),np.uint8)*255
```

```python
hist = cv2.calcHist([blank_w],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

cv2.imshow("blank_w", blank_w)
# cv2.imshow("blank_b", blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist = cv2.calcHist([blank_b],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# cv2.imshow("blank_w", blank_w)
cv2.imshow("blank_b", blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("image/old_men.jpg")
hist = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

cv2.imshow("old men", img)
# cv2.imshow("blank_b", blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("image/vk.jpg")
img = cv2.resize(img,(500,600))

b,g,r = cv2.split(img)

plt.hist(b.ravel(),255,(0,255),color="blue")
plt.hist(g.ravel(),255,(0,255),color="green")
plt.hist(r.ravel(),255,(0,255),color="red")
plt.show()

cv2.imshow("vk", img)
cv2.imshow("vk_b", b)
cv2.imshow("vk_g", g)
cv2.imshow("vk_r", r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
